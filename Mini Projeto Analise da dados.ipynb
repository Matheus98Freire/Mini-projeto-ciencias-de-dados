# Mini Projeto de Análise de Dados com Pandas


### Qual o desafio?

Tratar dados de uma empresa de prestação de serviços fictícia utilizando o Pandas. 



### Importação de Módulos e arquivos


```python
import pandas as pd

funcionarios_df = pd.read_csv('CadastroFuncionarios.csv', sep=';', decimal=',')
clientes_df = pd.read_csv('CadastroClientes.csv', sep = ';', decimal =',')
servicos_df = pd.read_excel("BaseServiçosPrestados.xlsx")
#excluir colunas Estado civil e cargo
funcionarios_df = funcionarios_df.drop(["Estado Civil","Cargo"], axis = 1)
display(funcionarios_df)
display(clientes_df)
display(servicos_df)
```

### 1- Total da Folha Salarial

```python
funcionarios_df["Salário Total"] = funcionarios_df["Salario Base"] + funcionarios_df["Impostos"] + funcionarios_df["Beneficios"] + funcionarios_df["VT"] + funcionarios_df["VR"] 
print("Total da folha salarial mensal é de R$ {:,}".format(sum(funcionarios_df["Salário Total"])))
```

### 2- Faturamento Total da Empresa


```python
faturamento_df = servicos_df[["ID Cliente","Tempo Total de Contrato (Meses)"]].merge(clientes_df[["ID Cliente","Valor Contrato Mensal"]],on="ID Cliente")
faturamento_df["Faturamento Total"] = faturamento_df["Tempo Total de Contrato (Meses)"] * faturamento_df["Valor Contrato Mensal"]
print("O Faturamento total da empresa é de R$ {:,}".format(sum(faturamento_df["Faturamento Total"])))
```

### 3-  Calcular a % Funcionários que Fecharam Contrato

```python
funcionarios_fecharam_contrato = len(servicos_df["ID Funcionário"].unique())
qtde_funcionarios_total = len(funcionarios_df["ID Funcionário"])
print("Percentual de funcionários que fecharam contrato: {:.2%}".format(funcionarios_fecharam_contrato/qtde_funcionarios_total))
```

### 4- Calcular a Quantidade Contratos por Área

```python
contratos_area_df = servicos_df[["ID Funcionário"]].merge(funcionarios_df[["ID Funcionário","Area"]], on= "ID Funcionário")
contratos_area_qtde = contratos_area_df["Area"].value_counts()
print(contratos_area_qtde)
contratos_area_qtde.plot(kind= "barh")
```

### 5- Funcionários por Área

```python
funcionarios_area = funcionarios_df["Area"].value_counts()
print(funcionarios_area)
funcionarios_area.plot(kind= "bar",color= "red")
```

### 6- Ticket Médio Mensal


```python
ticket_medio = clientes_df["Valor Contrato Mensal"].mean()
print("O ticket médio mensal foi: R${:,.2f}".format(ticket_medio))
```
